```python
import pysal as ps
import numpy as np
```

## Reading dbf files

```python
dbf = ps.open("data/south.dbf")
```

```python
dbf.header
```

```python
HR90 = dbf.by_col('HR90')
```

```python
HR90
```

```python
type(HR90)
```

```python
type(HR90[0])
```

Note that the reader determines (or tries to guess) the data type upon read. In this case we have a python list returned with each element being a float.

We can convert this to a numpy array for subsequent numerical analysis.

```python
HR90 = np.array(dbf.by_col("HR90"))
```

```python
HR90
```

```python
First_row = dbf.by_row[0]
```

```python
First_row
```

## Reading Shape Files

```python
south = ps.open("data/south.shp")
```

```python
south.header
```

```python
shps = []
for shp in south:
    shps.append(shp)
```

```python
len(shps)
```

```python
shp0 = shps[0]
```

```python
type(shp0)
```

We won't have much cause for reading the shapefiles for manipulating individual shapes, but in case you do, this is how.

## Reading csv files

```python
f = ps.open("data/usjoin.csv",'r')
```

To determine what is in the file, check the `header` attribute on the file object:

```python
f.header
```

Ok, lets pull in the name variable to see what we have

```python
name = f.by_col('Name')
```

```python
name
```

Now obtain per capita incomes in 1929 which is in the column associated with `1929`

```python
y1929 = f.by_col('1929')
```

```python
y1929
```

And now 2009

```python
y2009 = f.by_col("2009")
```

```python
y2009
```

These are read into regular Python lists which are not particularly well suited to efficient data analysis. So let's convert them to numpy arrays

```python
y2009 = np.array(y2009)
```

```python
y2009
```

Much better. But pulling these in and converting them a column at a time is tedious and error prone. So we will do all of this in a list comprehension.

```python
Y = np.array( [ f.by_col(str(year)) for year in range(1929,2010) ] ) * 1.0
```

```python
Y.shape
```

```python
Y = Y.transpose()
```

```python
Y.shape
```
