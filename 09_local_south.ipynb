## Binding Attribute Data to the Map

````python
import json

f = open(r'data/south.json')
q = json.load(f)
f.close()
````

````python
q.keys()
````

````python
features = q['features']
len(features)
````

````python
import pandas as pd
indices = []
values = []
for feature in features:
    indices.append(str(feature['properties']['FIPS']))
    values.append(feature['properties']['HR80'])
````

````python
df = pd.DataFrame({'HR80': values,
                   'FIPS': indices} )
````

````python
y = np.array(df.HR80.tolist())
````

````python
import numpy as np
import pysal as ps
w = ps.queen_from_shapefile('data/south.shp')
w.transform = 'r'
````

````python
y = np.array(df['HR80'].tolist())
````

````python
li_hr80 = ps.Moran_Local(y,w)
````

````python
li_hr80.p_sim.shape
````

````python
sig01 = 1.* (li_hr80.p_sim<=0.01)
````

````python
sig01.sum()
````

````python
sig05 = 1.* (li_hr80.p_sim<=0.05)
````

````python
sig05.sum()
````

````python
df['sig01']  = sig01
df['sig05'] = sig05
````
