```python
import numpy as np
import pysal as ps
w = ps.queen_from_shapefile('data/south.shp')
w.transform = 'r'
```

```python
y = np.array(df['HR80'].tolist())
```

## Local Moran
$li_i = \frac{n-1}{\sum_j z_j^2}z_i \sum_j w_{i,j} z_j$

```python
li_hr80 = ps.Moran_Local(y,w)
```

```python
li_hr80.p_sim.shape
```

```python
sig01 = 1.* (li_hr80.p_sim<=0.01)
```

```python
sig01.sum()
```

```python
sig05 = 1.* (li_hr80.p_sim<=0.05)
```

```python
sig05.sum()
```

```python
df['sig01']  = sig01
df['sig05'] = sig05
```
